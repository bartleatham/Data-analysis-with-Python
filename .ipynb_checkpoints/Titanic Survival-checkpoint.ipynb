## The Titanic Survivors, what passenger demographics tell us

In this project I will delve in to the [Titanic Passenger demographics](https://www.kaggle.com/c/titanic/data) as provided by Kaggle.com.  I will be investigating the relationships between passenger survival and various demographic variables in an attempt to draw a tentative conclusion on whether said demographic variables affected the chances of surival in the Titanic tragedy.



First import the necessary modules, and read the relevant data into a Pandas dataframe.  I am dropping columns that I dont find relevant to my study, and renaming Sex category to Gender to ensure a G rating for our younger viewers.

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
style.use('ggplot')

df = pd.read_csv('titanic-data.csv')
df.drop((['PassengerId', 'Ticket', 'Fare', 'Cabin', 'Embarked']), axis=1, inplace=True)
df.rename(columns={'Sex':'Gender'}, inplace=True)
```

**Wrangle the Data!**  
Check for null entries, and for possible duplicate rows.

```python
print("The DataFrame info is:")
df.info()
print("")
print("Categories with NULL data:")
print(df.isnull().sum())
print("")
print('There are {} duplicate rows'.format(df.duplicated().sum()))
```

From the results of df.info(), we can see that the number of entries for Age is less than the number of entries for all other fields.  Further, we can see that Age has 177 null entries.  We will proceed with 2 dataframes, 1 with all entries for Age, including null values, and 1 with null Age entries removed so that null entries do not skew Age based results.  Also creating a new DataFrame column for 'Child (1)' or 'Adult (0)', assuming that in that era a person 16 or older would likely be considered an adult. 

```python
df['Child'] = (df['Age'] < 16).astype(int)
df_age = df.dropna()
df_age.info()
df_age.isnull().sum()
```

**Get some High Level statistics on the passengers.**

```python
print('The Total Passengers with demographic data was {}.'.format(len(df.index)))
print('The Total Surviving Passenger count was {}.'.format((df['Survived']).sum()))
survivor_percentage= '{:.2f}'.format((df['Survived']).mean()*100)
print('The Percentage of Survivors was {}%.'.format(survivor_percentage))
print("")
print('The male and female counts are:')
gender = df['Gender'].value_counts() 
print(gender)
gender.plot.bar()
plt.show()

print('The child(1) and adult(0) counts are:')
child = df['Child'].value_counts() 
print(child)
child.plot.bar()
plt.show()
```

**Explore the Data!**     
Questions of Interest:  
1- Did the Socio-economic status (SES) affect the survival rate?  
2- Did age affect the survival rate?  
3- Did 'women and children first' play out as expected?  

**1- Did the Socio-economic status (SES) affect the survival rate?**  
We will assume that the passenger class (Pclass) is representative of the SES of the passenger:  
Pclass 1 -> high SES  
Pclass 2 -> mid SES  
Pclass 3 -> low SES

```python
survivors_by_class = df.groupby('Pclass')['Survived'].mean()
first_class = '{:.2f}'.format((survivors_by_class[1])*100)
second_class = '{:.2f}'.format((survivors_by_class[2])*100)
third_class = '{:.2f}'.format((survivors_by_class[3])*100)
print('The First Class Survival Rate was {}%.'.format(first_class))
print('The Second Class Survival Rate was {}%.'.format(second_class))
print('The Third Class Survival Rate was {}%.'.format(third_class))

survivors_by_class.plot.bar()
plt.xlabel('Passenger Class')
plt.ylabel('Survival Probability')
plt.title('Passenger Class vs. Survival Probability')
plt.show()

```

It is clear from the calculated survival rate percentages, and the graphical visualization that the passenger class, and thus the assumed SES did have a significant impact on the chances of survival, with those in higher status having a greater chance of survival.
From the data it is unclear how the SES impact is manifested.  It is safe to assume that the higher the passenger class, the closer the cabins are to the upper decks, and thus to the lifeboats.  It is possible that a passenger of higher SES, even if located in a lower passenger class cabin could have been given preferential treatment in evacuation by their fellow passengers.  Again, we do not have the proper data to substantiate these hypothesis.

**2- Did age have an affect on survival rate?**  
Will use df_age DataFrame with Nan removed.  
Will seperate children (< 16yrs) from adults (>= 16yrs).

```python
print('The Total Boarding Passengers with Age data was {}.'.format(len(df_age.index)))
print('The Total Surviving Passenger count with Age data was {}.'.format((df_age['Survived']).sum()))

survivors_by_age = df_age.groupby('Child')['Survived'].mean()
adult_survival_pct = '{:.2f}'.format((survivors_by_age[0])*100)
child_survival_pct = '{:.2f}'.format((survivors_by_age[1])*100)
print('The Percentage of Adults aged 16 and over that survived was {}%.'.format(adult_survival_pct))
print('The Percentage of Children aged 15 and under that survived was {}%.'.format(child_survival_pct))


plt.xlabel('Passenger Age (yrs)')
plt.title('Ages of all Passengers and Surviving Passengers')
df_age['Age'].plot.hist(bins=25, label = 'Boarding Passenger Ages')
df_age_surv = df_age[(df_age['Survived'] == 1)]
df_age_surv['Age'].plot.hist(bins=25, label = 'Surviving Passenger Ages')
plt.legend()
plt.show()
```

From the survival rate percentages calculated above, we can see that children age 15 and younger did have a higher likely hood of survival, 57% vs 38%.

**3- Did 'women and children first' apply?**


```python
survivors_by_gender_and_age = df_age.groupby((['Gender', 'Child']),as_index=True)['Survived'].mean()

print(survivors_by_gender_and_age)

survivors_by_gender_and_age.plot.bar()
plt.xlabel('Gender, Child (1=Child, 0=Adult)')
plt.ylabel('Survival Probability')
plt.title('Gender/Age vs. Survival Probability')
plt.show()
```

Again, from the survival percentages calculated above, we can see that being a female and/or a child did have a noticable affect on survival probability.  Being an adult male on board the Titanic led to a dismal outcome for most of them.  Chivalry aboard the Titanic was not dead!  Although most of the chivalrous did end up dead.

## Conclusion: 
Although the data provided for this exercise is incomplete, and not a thorough representation of the entire population of passengers aboard the Titanic, we can make some general observations given the data at hand.
Being a lower class adult male was not a desireable demographic to belong to, these men showed the lowest survival probability.  Being an adult female resulted in the highest survival probability.
